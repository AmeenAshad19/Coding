```python
# Recursive:
def catalan(n):
	if n <= 1:
		return 1
	res = 0
	for i in range(n):
		res += catalan(i) * catalan(n-i-1)

	return res

for i in range(10):
	print(catalan(i), end=" ")
```

```python
# DP:
def catalan(n):
	if (n == 0 or n == 1):
		return 1

	catalan = [0]*(n+1)

	catalan[0] = 1
	catalan[1] = 1

	for i in range(2, n + 1):
		for j in range(i):
			catalan[i] += catalan[j] * catalan[i-j-1]

	return catalan[n]

for i in range(10):
	print(catalan(i), end=" ")
```

```python
def minOperation(k):
	dp = [0] * (k + 1)

	for i in range(1, k + 1):
		dp[i] = dp[i - 1] + 1
		if (i % 2 == 0):
			dp[i]= min(dp[i], dp[i // 2] + 1)

	return dp[k]

if __name__ == '__main__':
	
	k = 12
	
	print(minOperation(k))
```
